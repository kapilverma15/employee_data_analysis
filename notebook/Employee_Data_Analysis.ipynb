```python
import pandas as pd
import numpy as np

# Same output every time
np.random.seed(42)

# Total Records
n = 1000

# Sample Data
names = ["Amit","Rahul","Kapil","Ankit","Rohit","Priya","Neha","Pooja","Anjali","Vikas"]
departments = ["HR","IT","Finance","Sales","Marketing","Admin"]
cities = ["Delhi","Noida","Gurugram","Faridabad","Jaipur"]
genders = ["Male","Female"]

# Create DataFrame
df = pd.DataFrame({
    "Employee_ID": range(1001,1001+n),
    "Name": np.random.choice(names,n),
    "Department": np.random.choice(departments,n),
    "Gender": np.random.choice(genders,n),
    "Age": np.random.randint(21,60,n),
    "Salary": np.random.randint(25000,120001,n),
    "Experience": np.random.randint(0,20,n),
    "City": np.random.choice(cities,n),
    "Performance": np.random.randint(1,6,n)
})

# Display first 5 rows
df.head()
```

```python
# Create a copy of the dataset
df_clean = df.copy()

print("Working copy created successfully!")
```

```python
# Artificial Missing Values

df_clean.loc[10, "Salary"] = np.nan
df_clean.loc[25, "Age"] = np.nan
df_clean.loc[50, "Department"] = np.nan
df_clean.loc[100, "City"] = np.nan

print("Missing values inserted successfully!")
```

```python
df_clean.isnull().sum()
```

```python
# Fill Missing Values

# Numeric Columns
df_clean["Salary"] = df_clean["Salary"].fillna(df_clean["Salary"].mean())
df_clean["Age"] = df_clean["Age"].fillna(df_clean["Age"].median())

# Categorical Columns
df_clean["Department"] = df_clean["Department"].fillna("Unknown")
df_clean["City"] = df_clean["City"].fillna("Unknown")

print("Missing values handled successfully!")
```

```python
performance = df.groupby("Department", as_index=False)["Performance"].mean()

performance.plot(
    x="Department",
    y="Performance",
    kind="bar",
    figsize=(8,5),
    legend=False,
    title="Average Performance by Department"
)

plt.tight_layout()
plt.savefig("Bar_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
salary_exp = df.groupby("Experience", as_index=False)["Salary"].mean()

salary_exp.plot(
    x="Experience",
    y="Salary",
    kind="line",
    marker="o",
    figsize=(10,5),
    title="Average Salary by Experience"
)

plt.grid(True)
plt.tight_layout()

# Save chart
plt.savefig("line_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```
